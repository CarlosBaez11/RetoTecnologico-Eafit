Es importante verificar si se puede deducir la faltra de cuadrante a partir de la causal de mora que presenta el cliente. En la tabla siguiete podemos ver que la gran mayoría pertenece a la cetegoría sin causal de mora, por tanto, es necesario establecer otros mecanismos de seguimiento, dado que se está perdiendo toda la información acerca de la causa del problema de las personas sin oficio que están en mora

```python
cuadrante=datos[datos["Ocupacion"]=="SIN OFICIO"]
cuadrante=cuadrante[cuadrante["Cuadrante"]=="SIN CUADRANTE"]
```

```python
cuadrante.groupby(["Cuadrante", "Descripcion Causal"])["Cuadrante"].aggregate(["count"]).sort_values(by="count", ascending=False)
```

# Hagamos un análisis de la descripción causal

Al hacer un conteo de la descripción causal en todo el dataset podemos observar que la gran mayoría de causas está indocumentada, o es generada por descuido

```python
abogados=datos.groupby(["Descripcion Causal"])["Descripcion Causal"].aggregate(["count"])
abogados.sort_values(by="count", ascending=False)
```

Al comparar los resultados de la tabla de **conteo por segmentación causal** con la tabla de **segmentación causal por cuadrante** podemos observar que la gran mayoría de personas que no tienen  cuadrante, tampoco especifican una descripción causal, por tanto hay una fuerte correlación entre las personas que no tienen causal de mora y no pertenecen a ningún cuadrante

Ahora para determinar la cantidad de dinero que deben las personas con respecto a su cuadrante para verificar si es recomendable crear un foco en las personas que no tienen datos, con el propósito de intentar hallar una relación entre ellos

```python
endeudamiento_cuadrante=datos.groupby(["Cuadrante"]).aggregate({"Endeudamiento":"sum", "Dias Mora":"mean"}).sort_values(by="Endeudamiento", ascending=False)
endeudamiento_cuadrante
```

Como podemos ver en la imagen anterior, la mayoría de dinero en mora se encuentra entre las personas de los cuadrantes **TIENE VOLUNTAD Y TIENE CAPACIDAD REDUCIDA, TIENE VOLUNTAD Y TIENE CAPACIDAD, SIN CUADRANTE**
