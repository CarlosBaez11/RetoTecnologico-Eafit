```python
cuadrante=datos[datos["Ocupacion"]=="SIN OFICIO"]
cuadrante=cuadrante[cuadrante["Cuadrante"]=="SIN CUADRANTE"]
```

```python
cuadrante.groupby(["Cuadrante", "Descripcion Causal"])["Cuadrante"].aggregate(["count"]).sort_values(by="count", ascending=False)
```

# Hagamos un anÃ¡lisis de la descripciÃ³n causal

Al hacer un conteo de la descripciÃ³n causal en todo el dataset podemos observar que la gran mayorÃ­a de causas estÃ¡ indocumentada, o es generada por descuido

```python
abogados=datos.groupby(["Descripcion Causal"])["Descripcion Causal"].aggregate(["count"])
abogados.sort_values(by="count", ascending=False)
```

Al comparar los resultados de la tabla de **conteo por segmentaciÃ³n causal** con la tabla de **segmentaciÃ³n causal por cuadrante** podemos observar que la gran mayorÃ­a de personas que no tienen  cuadrante, tampoco especifican una descripciÃ³n causal, por tanto hay una fuerte correlaciÃ³n entre las personas que no tienen causal de mora y no pertenecen a ningÃºn cuadrante

Ahora para determinar la cantidad de dinero que deben las personas con respecto a su cuadrante para verificar si es recomendable crear un foco en las personas que no tienen datos, con el propÃ³sito de intentar hallar una relaciÃ³n entre ellos

```python
endeudamiento_cuadrante=datos.groupby(["Cuadrante"]).aggregate({"Endeudamiento":"sum", "Dias Mora":"mean"}).sort_values(by="Endeudamiento", ascending=False)
endeudamiento_cuadrante
```

Como podemos ver en la imagen anterior, la mayorÃ­a de dinero en mora se encuentra entre las personas de los cuadrantes **TIENE VOLUNTAD Y TIENE CAPACIDAD REDUCIDA, TIENE VOLUNTAD Y TIENE CAPACIDAD, SIN CUADRANTE**
